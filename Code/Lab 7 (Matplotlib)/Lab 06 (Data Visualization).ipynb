## Setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

##  1️⃣ Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True)
plt.show()
```

2️⃣ Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

3️⃣ Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [45, 30, 15, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio
plt.show()
```

4️⃣ Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

5️⃣ Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

6️⃣ Subplots

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, np.sin(x)); axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, np.cos(x)); axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, np.tan(x)); axs[1, 0].set_title('tan(x)')
axs[1, 1].plot(x, np.exp(x)); axs[1, 1].set_title('exp(x)')
plt.tight_layout()
plt.show()
```

7️⃣ Plot Styles and Themes


```python
plt.style.use('ggplot')  # Try: 'seaborn', 'fivethirtyeight', 'bmh', 'dark_background', etc.

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='Sin', linewidth=2)
plt.plot(x, np.cos(x), label='Cos', linewidth=2)
plt.title("Styled Plot")
plt.legend()
plt.show()
```

🔹 Twin Axes

```python
x = np.arange(1, 11)
y1 = x**2
y2 = np.log(x)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', label='Square')
ax2.plot(x, y2, 'b--', label='Log')
ax1.set_xlabel('X')
ax1.set_ylabel('Square', color='g')
ax2.set_ylabel('Log', color='b')
plt.title("Twin Axes")
plt.show()
```

🔥 Advanced Matplotlib Topics

1️⃣ Interactive Plots with Widgets (Jupyter)

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

x = np.linspace(0, 10, 100)

def update(freq=1.0, amp=1.0):
    y = amp * np.sin(freq * x)
    plt.figure(figsize=(8, 4))
    plt.plot(x, y)
    plt.ylim(-2, 2)
    plt.title(f"Sin with freq={freq} and amp={amp}")
    plt.show()

interact(update, freq=(0.5, 5.0, 0.1), amp=(0.5, 2.0, 0.1));
```

2️⃣ 3D Plots with Axes3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Surface Plot")
plt.show()
```

3️⃣ Heatmaps

```python
import seaborn as sns
data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

4️⃣ Animation

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 200)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i / 10.0))  # update data
    return line,

ani = animation.FuncAnimation(fig, animate, frames=100, interval=50, blit=True)
plt.show()
```

5️⃣ Multiple Y-Axes

```python
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = np.log(x + 1)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(x, y1, 'g-', label='x^2')
ax2.plot(x, y2, 'b--', label='log(x)')

ax1.set_ylabel('x^2', color='g')
ax2.set_ylabel('log(x)', color='b')
plt.title("Two Y-Axes")
plt.show()
```

6️⃣ Inset Plots

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y)
inset = fig.add_axes([0.6, 0.5, 0.25, 0.25])
inset.plot(x, y)
inset.set_title("Inset", fontsize=8)
plt.show()
```

7️⃣ Custom Colormaps

```python
from matplotlib import cm

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

plt.contourf(X, Y, Z, 20, cmap=cm.inferno)
plt.colorbar()
plt.title("Custom Colormap with Contour")
plt.show()
```

8️⃣ Seaborn Integration for Statistical Plots

```python
import seaborn as sns
tips = sns.load_dataset("tips")

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin Plot with Seaborn")
plt.show()
```

9️⃣ Polar Plots

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(5 * theta))

plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

🔟 Tight Layout, Padding, and Custom Sizes

### **Plot multiple curves**

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)       # First curve (sine wave)
y2 = np.cos(x)       # Second curve (cosine wave)
y3 = np.sin(x) * 0.5 # Third curve (damped sine wave)

# Create figure
# plt.figure(figsize=(10, 6))

# Plot multiple curves on the same graph
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.plot(x, y3, label='0.5*sin(x)', color='green', linestyle=':', linewidth=2)

# Add titles and labels
# plt.title("Multiple Plots on Same Graph", fontsize=14, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)

# Add legend and grid
# plt.legend(fontsize=10)
# plt.grid(True, linestyle='--', alpha=0.5)

# Adjust layout and show plot
# plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```
